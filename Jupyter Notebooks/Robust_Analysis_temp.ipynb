```python
import os
import sys
import pandas as pd
import numpy as np
import scanpy as sc
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import adjusted_rand_score
from MuCoST.mucost import training_model
from MuCoST.utils import mclust
from MuCoST.config import set_arg
```

### robust

```python
opt = set_arg()
arg = opt.parse_args(['--mode_his', 'noh'])
```

```python
# arg.n_domain = 7
sections_7 = ['151507', '151508', '151509', '151510', '151673', '151674', '151675', '151676']
# arg.n_domain = 5
sections_5 = ['151669', '151670', '151671', '151672']
temp = [1,0.7,0.5,0.3,0.2,0.1,0.09,0.08,0.07,0.06,0.05,0.04,0.03,0.02,0.01]
for j, t in enumerate(temp):
    arg.temp = t
    ARIS = {}
    for i, s in enumerate(sections_7):
        # reading dataset
        arg.n_domain = 7
        section_id = s
        input_dir = os.path.join('Data/1-DLPFC', section_id)
        adata = sc.read_visium(path=input_dir, count_file='filtered_feature_bc_matrix.h5')

        # reading ground truth
        Ann_df = pd.read_csv(os.path.join('Data/1-DLPFC_annotations', section_id+'_truth.txt'), sep='\t', header=None, index_col=0)
        Ann_df.columns = ['Ground Truth']
        adata.obs['Ground Truth'] = Ann_df.loc[adata.obs_names, 'Ground Truth']

        training_model(adata, arg)
        # clustering
        adata = mclust(adata, arg, refine=False)
        # computing ARI
        adata = adata[~pd.isnull(adata.obs['Ground Truth'])]
        obs_df = adata.obs.dropna()
        ARI = adjusted_rand_score(obs_df['mclust'], obs_df['Ground Truth'])
        ARIS[section_id] = ARI
    for i, s in enumerate(sections_5):
        # reading dataset
        arg.n_domain = 5
        section_id = s
        input_dir = os.path.join('Data/1-DLPFC', section_id)
        adata = sc.read_visium(path=input_dir, count_file='filtered_feature_bc_matrix.h5')

        # reading ground truth
        Ann_df = pd.read_csv(os.path.join('Data/1-DLPFC_annotations', section_id+'_truth.txt'), sep='\t', header=None, index_col=0)
        Ann_df.columns = ['Ground Truth']
        adata.obs['Ground Truth'] = Ann_df.loc[adata.obs_names, 'Ground Truth']

        training_model(adata, arg)
        # clustering
        adata = mclust(adata, arg, refine=False)
        # computing ARI
        adata = adata[~pd.isnull(adata.obs['Ground Truth'])]
        obs_df = adata.obs.dropna()
        ARI = adjusted_rand_score(obs_df['mclust'], obs_df['Ground Truth'])
        ARIS[section_id] = ARI
    import json
    with open('report/temp-'+str(t)+'-results.txt', 'w', encoding='utf-8') as f:
        json.dump(ARIS, f)
    f.close()
```
