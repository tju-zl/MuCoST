```python
import os
import sys
import pandas as pd
import numpy as np
import scanpy as sc
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import adjusted_rand_score
from MuCoST.mucost import training_model
from MuCoST.utils import mclust
from MuCoST.config import set_arg
```

```python
opt = set_arg()
arg = opt.parse_args(['--mode_his', 'noh'])
arg.temp = 0.05
arg
```

```python
# arg.n_domain = 7
sections_7 = ['151507', '151508', '151509', '151510', '151673', '151674', '151675', '151676']
# arg.n_domain = 5
sections_5 = ['151669', '151670', '151671', '151672']
ARIS = {}
NMIS = {}
HS = {}
CS = {}
V05 = {}
V15 = {}
```

```python
for i, s in enumerate(sections_7):
    # reading dataset
    arg.n_domain = 7
    section_id = s
    print(s)
    input_dir = os.path.join('Data/1-DLPFC', section_id)
    adata = sc.read_visium(path=input_dir, count_file='filtered_feature_bc_matrix.h5')
    # reading ground truth
    Ann_df = pd.read_csv(os.path.join('Data/1-DLPFC_annotations', section_id+'_truth.txt'), sep='\t', header=None, index_col=0)
    Ann_df.columns = ['Ground Truth']
    adata.obs['Ground Truth'] = Ann_df.loc[adata.obs_names, 'Ground Truth']

    training_model(adata, arg)
    # clustering
    adata = mclust(adata, arg, refine=False)
    # computing ARI
    adata = adata[~pd.isnull(adata.obs['Ground Truth'])]
    obs_df = adata.obs.dropna()
    ARI = adjusted_rand_score(obs_df['mclust'], obs_df['Ground Truth'])
    ARIS[section_id] = ARI
    print(ARI)
    from sklearn.metrics.cluster import normalized_mutual_info_score
    nmi = normalized_mutual_info_score(obs_df['mclust'], obs_df['Ground Truth'])
    NMIS[section_id] = nmi
    from sklearn.metrics import homogeneity_score
    home = homogeneity_score(obs_df['Ground Truth'], obs_df['mclust'])
    HS[section_id] = home
    from sklearn.metrics import completeness_score
    completeness = completeness_score(obs_df['Ground Truth'], obs_df['mclust'])
    CS[section_id] = completeness
    from sklearn.metrics import v_measure_score
    v_05 = v_measure_score(obs_df['Ground Truth'],obs_df['mclust'],beta=0.5)
    V05[section_id]=v_05
    v_15 = v_measure_score(obs_df['Ground Truth'],obs_df['mclust'],beta=1.5)
    V15[section_id]=v_15
for i, s in enumerate(sections_5):
    # reading dataset
    arg.n_domain = 5
    section_id = s
    print(s)
    input_dir = os.path.join('Data/1-DLPFC', section_id)
    adata = sc.read_visium(path=input_dir, count_file='filtered_feature_bc_matrix.h5')
    # reading ground truth
    Ann_df = pd.read_csv(os.path.join('Data/1-DLPFC_annotations', section_id+'_truth.txt'), sep='\t', header=None, index_col=0)
    Ann_df.columns = ['Ground Truth']
    adata.obs['Ground Truth'] = Ann_df.loc[adata.obs_names, 'Ground Truth']

    training_model(adata, arg)
    # clustering
    adata = mclust(adata, arg, refine=False)
    # computing ARI
    adata = adata[~pd.isnull(adata.obs['Ground Truth'])]
    obs_df = adata.obs.dropna()
    ARI = adjusted_rand_score(obs_df['mclust'], obs_df['Ground Truth'])
    ARIS[section_id] = ARI
    print(ARI)
    from sklearn.metrics.cluster import normalized_mutual_info_score
    nmi = normalized_mutual_info_score(obs_df['mclust'], obs_df['Ground Truth'])
    NMIS[section_id] = nmi
    from sklearn.metrics import homogeneity_score
    home = homogeneity_score(obs_df['Ground Truth'], obs_df['mclust'])
    HS[section_id] = home
    from sklearn.metrics import completeness_score
    completeness = completeness_score(obs_df['Ground Truth'], obs_df['mclust'])
    CS[section_id] = completeness
    from sklearn.metrics import v_measure_score
    v_05 = v_measure_score(obs_df['Ground Truth'],obs_df['mclust'],beta=0.5)
    V05[section_id]=v_05
    v_15 = v_measure_score(obs_df['Ground Truth'],obs_df['mclust'],beta=1.5)
    V15[section_id]=v_15
```

```python
dir = 'report/MuCoST-'
import json
with open(dir+'ari-results.txt', 'w', encoding='utf-8') as f:
    json.dump(ARIS, f)
f.close()
with open(dir+'nmi-results.txt', 'w', encoding='utf-8') as f:
    json.dump(NMIS, f)
f.close()
with open(dir+'hs-results.txt', 'w', encoding='utf-8') as f:
    json.dump(HS, f)
f.close()
with open(dir+'cs-results.txt', 'w', encoding='utf-8') as f:
    json.dump(CS, f)
f.close()
with open(dir+'v05-results.txt', 'w', encoding='utf-8') as f:
    json.dump(V05, f)
f.close()
with open(dir+'v15-results.txt', 'w', encoding='utf-8') as f:
    json.dump(V15, f)
f.close()
print('end')
```
